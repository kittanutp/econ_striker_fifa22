```python
import pandas as pd
df = pd.read_csv('players_22.csv', low_memory = True)
print(list(df))
df.head()
```

```python
df.describe()
```

```python
df1 = pd.DataFrame(df, columns= ['long_name','value_eur','age', 'pace', 'shooting']).query('pace >= 75 & shooting >= 80 & age <= 28')
```

```python
df1['total_stat'] = (df1['pace']+df1['shooting'])
df1['s/p']= df1['total_stat']/df1['value_eur']
print(df1.sort_values('s/p', ascending=False).head())
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='value_eur', y='total_stat', data=df1)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Total Stat,Value)", "(@total_stat, @value_eur)"),
    ("Name", "@long_name")]
)

p = figure(title="Fifa 2022_Econ_Striker", x_axis_label='Total Stat', y_axis_label='Value in Euro', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('total_stat', 'value_eur', size=10, source=df1)
show(p)
```
